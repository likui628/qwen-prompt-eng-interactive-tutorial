## Setup

Run the following setup cell to load your API key and establish the `get_completion` helper function.

```python
import os
from openai import OpenAI

# Retrieve the MODEL_NAME variable from the IPython store
%store -r MODEL_NAME

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"), base_url='https://dashscope.aliyuncs.com/compatible-mode/v1')

# Note that we changed max_tokens to 4K just for this lesson to allow for longer completions in the exercises
def get_completion(prompt: str, system_prompt=""):
    messages = []
    if system_prompt:
        messages.append({"role": "system", "content": system_prompt})
    messages.append({"role": "user", "content": prompt})
    
    response = client.chat.completions.create(
        model=MODEL_NAME,
        max_tokens=4000,
        temperature=0.0,
        messages=messages
    )
    return response.choices[0].message.content
```

## Lesson

**Qwen responds best to clear and direct instructions.**

Think of Qwen like any other human that is new to the job. **Qwen has no context** on what to do aside from what you literally tell it. Just as when you instruct a human for the first time on a task, the more you explain exactly what you want in a straightforward manner to Qwen, the better and more accurate Qwen's response will be."				
				
When in doubt, follow the **Golden Rule of Clear Prompting**:
- Show your prompt to a colleague or friend and have them follow the instructions themselves to see if they can produce the result you want. If they're confused, Qwen's confused.

### Examples

Let's take a task like writing poetry. (Ignore any syllable mismatch - LLMs aren't great at counting syllables yet.)

```python
# Prompt
PROMPT = "Write a haiku about robots."

# Print Qwen's response
print(get_completion(PROMPT))
```

This haiku is nice enough, but users may want Qwen to go directly into the poem without the "Here is a haiku" preamble.

How do we achieve that? We **ask for it**!

```python
# Prompt
PROMPT = "Write a haiku about robots. Skip the preamble; go straight into the poem."

# Print Qwen's response
print(get_completion(PROMPT))
```

Here's another example. Let's ask Qwen who's the best basketball player of all time. You can see below that while Qwen lists a few names, **it doesn't respond with a definitive "best"**.

```python
# Prompt
PROMPT = "Who is the best basketball player of all time?"

# Print Qwen's response
print(get_completion(PROMPT))
```

Can we get Qwen to make up its mind and decide on a best player? Yes! Just ask!

```python
# Prompt
PROMPT = "Who is the best basketball player of all time? Yes, there are differing opinions, but if you absolutely had to pick one player, who would it be?"

# Print Qwen's response
print(get_completion(PROMPT))
```

### Exercise 2.1 - Spanish
Modify the `SYSTEM_PROMPT` to make Qwen output its answer in Spanish.

```python
# System prompt - this is the only field you should chnage
SYSTEM_PROMPT = "[Replace this text]"

# Prompt
PROMPT = "Hello Qwen, how are you?"

# Get Qwen's response
response = get_completion(PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return "hola" in text.lower()

# Print Qwen's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```

### Exercise 2.3 - Write a Story

Modify the `PROMPT` so that Qwen responds with as long a response as you can muster. If your answer is **over 800 words**, Qwen's response will be graded as correct.

```python
# Prompt - this is the only field you should change
PROMPT = "[Replace this text]"

# Get Qwen's response
response = get_completion(PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    trimmed = text.strip()
    words = len(trimmed.split())
    return words >= 800

# Print Qwen's response and the corresponding grade
print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", grade_exercise(response))
```
